To create this chatbot i found a nice dataset and kaggle https://www.kaggle.com/kausr25/chatterbotenglish?select=ai.yml

First we will download all the data as <i>.yml</i> files, and extract the data in a jupyter notebook


```python
import os 
import yaml
import nltk
import tensorflow
dir_path = 'raw_data'
```

```python
file_prefix = './raw_data'
```

```python
file_path = os.listdir(f'./{dir_path}')
```

```python
questions = list()
labels = list()
answers = list()
```

```python
def tokennize_word(word: str):
    try:
        nltk.word_tokenize(word)
    except Exception as e:
        print(f'Failed for word {word} with {str(e)}')
        
```

```python
for file in file_path:
    stream = open(f'{file_prefix}/{file}', 'rb')
    data = yaml.safe_load(stream)
    conversations = data['conversations']
    labels.append(file.split('.')[0])
    #generic solution, that concats multiple answers to the same question
    #some of the data is repeated, but that is okay for know
    for con in conversations:
        if len( con ) > 2 :
            questions.append(con[0])
            replies = con[ 1 : ]
            ans = ''
            for rep in replies:
                ans += ' ' + rep
            if ans not in answers:    
                answers.append( ans )
        elif len( con )> 1:
            questions.append(con[0])
            if con[1] not in answers:
                answers.append(con[1])
            
            
    
    
```

```python
print(f'questions: {len(questions)} | answers {len(answers)}')
```
